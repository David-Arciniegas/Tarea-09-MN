```python
import numpy as np
import matplotlib.pyplot as plt

def plot_system(a, b):
    try:
        # Extraer los coeficientes
        A = np.array([a[0][:2], a[1][:2]])
        B = np.array([b[0], b[1]])

        # Resolver el sistema
        sol = np.linalg.solve(A, B)

        # Crear los datos para la gráfica
        x = np.linspace(-10, 10, 400)
        y1 = (-a[0][0]*x + b[0]) / a[0][1]
        y2 = (-a[1][0]*x + b[1]) / a[1][1]

        # Graficar las rectas
        plt.figure()
        plt.plot(x, y1, label=f'{a[0][0]}x + {a[0][1]}y = {b[0]}')
        plt.plot(x, y2, label=f'{a[1][0]}x + {a[1][1]}y = {b[1]}')
        plt.scatter(sol[0], sol[1], color='red', zorder=5)
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Solución gráfica del sistema de ecuaciones')
        plt.grid(True)
        plt.axhline(0, color='black',linewidth=0.5)
        plt.axvline(0, color='black',linewidth=0.5)
        plt.show()

        return sol

    except np.linalg.LinAlgError:
        # Crear los datos para la gráfica
        x = np.linspace(-10, 10, 400)
        y1 = (-a[0][0]*x + b[0]) / a[0][1]
        y2 = (-a[1][0]*x + b[1]) / a[1][1]

        # Graficar las rectas
        plt.figure()
        plt.plot(x, y1, label=f'{a[0][0]}x + {a[0][1]}y = {b[0]}')
        plt.plot(x, y2, label=f'{a[1][0]}x + {a[1][1]}y = {b[1]}')
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Solución gráfica del sistema de ecuaciones')
        plt.grid(True)
        plt.axhline(0, color='black',linewidth=0.5)
        plt.axvline(0, color='black',linewidth=0.5)
        plt.show()

        return "Sistema sin solución única (matriz singular)"


```

```python
# matriz
matriz= ([[2.0, 1], [1, 1],[1,-3]], [-1, 2, 5])

solucion = plot_system(*matriz)
print(f"Solución de la matriz es: {solucion}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_system(a, b, z_value=0):
    try:
        # Crear los datos para la gráfica
        x = np.linspace(-10, 10, 400)

        # Calcular y para las ecuaciones con z = 0
        y1 = (b[0] - a[0][2] * z_value - a[0][0] * x) / a[0][1]
        y2 = (b[1] - a[1][2] * z_value - a[1][0] * x) / a[1][1]

        # Graficar las rectas
        plt.figure(figsize=(8, 8))
        plt.plot(x, y1, label=f'{a[0][0]}x + {a[0][1]}y + {a[0][2]}z = {b[0]} (z = {z_value})')
        plt.plot(x, y2, label=f'{a[1][0]}x + {a[1][1]}y + {a[1][2]}z = {b[1]} (z = {z_value})')
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Solución gráfica del sistema de ecuaciones en 2D (con z = 0)')
        plt.grid(True)
        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        plt.show()

        return "Solución gráfica generada"

    except np.linalg.LinAlgError:
        return "Sistema sin solución única (matriz singular)"

```

```python
import numpy as np

def gaussian_elimination(A, b, round_digits=None):
    """
    Solves the system of linear equations Ax = b using Gaussian elimination with back substitution.
    If round_digits is specified, rounds the intermediate results to the given number of digits.
    """
    A = A.astype(float)
    b = b.astype(float)

    if round_digits:
        A = np.round(A, round_digits)
        b = np.round(b, round_digits)

    n = len(b)

    # Forward elimination with partial pivoting
    for k in range(n):
        # Find the k-th pivot
        max_index = np.argmax(np.abs(A[k:n, k])) + k
        if A[max_index, k] == 0:
            raise ValueError("Matrix is singular or nearly singular")
        # Swap rows
        if max_index != k:
            A[[k, max_index]] = A[[max_index, k]]
            b[[k, max_index]] = b[[max_index, k]]

        for i in range(k+1, n):
            factor = A[i, k] / A[k, k]
            for j in range(k, n):
                A[i, j] -= factor * A[k, j]
                if round_digits:
                    A[i, j] = round(A[i, j], round_digits)
            b[i] -= factor * b[k]
            if round_digits:
                b[i] = round(b[i], round_digits)

    # Back substitution
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        sum_ax = 0
        for j in range(i+1, n):
            sum_ax += A[i, j] * x[j]
        x[i] = (b[i] - sum_ax) / A[i, i]
        if round_digits:
            x[i] = round(x[i], round_digits)

    return x

```

```python
import numpy as np

def gaussian_elimination2(A, b, dtype=np.float32, round_digits=None):
    """
    Solves the system of linear equations Ax = b using Gaussian elimination with back substitution.
    If dtype is specified, uses that numerical type for the computation.
    If round_digits is specified, rounds the intermediate results to the given number of digits.
    """
    A = A.astype(dtype)
    b = b.astype(dtype)

    if round_digits:
        A = np.round(A, round_digits)
        b = np.round(b, round_digits)

    n = len(b)
    row_swaps = []

    # Forward elimination with partial pivoting
    for k in range(n):
        # Find the k-th pivot
        max_index = np.argmax(np.abs(A[k:n, k])) + k
        if A[max_index, k] == 0:
            raise ValueError("Matrix is singular or nearly singular")
        # Swap rows
        if max_index != k:
            A[[k, max_index]] = A[[max_index, k]]
            b[[k, max_index]] = b[[max_index, k]]
            row_swaps.append((k, max_index))

        for i in range(k+1, n):
            factor = A[i, k] / A[k, k]
            for j in range(k, n):
                A[i, j] -= factor * A[k, j]
                if round_digits:
                    A[i, j] = round(A[i, j], round_digits)
            b[i] -= factor * b[k]
            if round_digits:
                b[i] = round(b[i], round_digits)

    # Back substitution
    x = np.zeros(n, dtype=dtype)
    for i in range(n-1, -1, -1):
        sum_ax = 0
        for j in range(i+1, n):
            sum_ax += A[i, j] * x[j]
        x[i] = (b[i] - sum_ax) / A[i, i]
        if round_digits:
            x[i] = round(x[i], round_digits)

    return x, row_swaps


```

```python
import numpy as np

def gauss_jordan(A, b, dtype=np.float32, round_digits=None):
    """
    Solves the system of linear equations Ax = b using the Gauss-Jordan elimination method.
    If dtype is specified, uses that numerical type for the computation.
    If round_digits is specified, rounds the intermediate results to the given number of digits.
    """
    A = A.astype(dtype)
    b = b.astype(dtype)

    if round_digits:
        A = np.round(A, round_digits)
        b = np.round(b, round_digits)

    n = len(b)
    row_swaps = []

    # Forward elimination with partial pivoting
    for k in range(n):
        # Find the k-th pivot
        max_index = np.argmax(np.abs(A[k:n, k])) + k
        if A[max_index, k] == 0:
            raise ValueError("Matrix is singular or nearly singular")
        # Swap rows
        if max_index != k:
            A[[k, max_index]] = A[[max_index, k]]
            b[[k, max_index]] = b[[max_index, k]]
            row_swaps.append((k, max_index))

        # Make the diagonal contain all 1's
        b[k] /= A[k, k]
        A[k, :] /= A[k, k]

        # Eliminate all other entries in the k-th column
        for i in range(n):
            if i != k:
                factor = A[i, k]
                A[i, :] -= factor * A[k, :]
                b[i] -= factor * b[k]
                if round_digits:
                    A[i, :] = np.round(A[i, :], round_digits)
                    b[i] = round(b[i], round_digits)

    return b, row_swaps
```

```python
# Ejercicio 4a con precisión de 32 bits
A4a = np.array([[0.25, 0.2, 0.167], [0.333, 0.25, 0.2], [0.5, 1, 2]], dtype=np.float32)
b4a = np.array([9, 8, 8], dtype=np.float32)

sol_4a, swaps_4a = gauss_jordan(A4a, b4a, dtype=np.float32)
print("Solución 7a:", sol_4a)
print("Intercambios de fila 7a:", swaps_4a)

A4b = np.array([[3.333, 15920, -10.333], [2.222, 16.71, 9.612], [1.5611, 5.1791, 1.6852]], dtype=np.float32)
b4b = np.array([15913, 28.544, 8.4254], dtype=np.float32)

sol_4b, swaps_4b = gauss_jordan(A4b, b4b, dtype=np.float32)
print("Solución 7b:", sol_4b)
print("Intercambios de fila 7b:", swaps_4b)

A4c = np.array([[1, 1/2, 1/3, 1/4], [1/2, 1/3, 1/4, 1/5], [1/3, 1/4, 1/5, 1/6], [1/4, 1/5, 1/6, 1/7]], dtype=np.float32)
b4c = np.array([1/6, 1/7, 1/8,1/9], dtype=np.float32)

sol_4c, swaps_4c = gauss_jordan(A4c, b4c, dtype=np.float32)
print("Solución 7c:", sol_4c)
print("Intercambios de fila 7c:", swaps_4c)

A4d = np.array([[2, 1, -1, 1, -3], [1, 0, 2, -1, 1], [0, -2, -1, 1, -1], [3, 1, -4, 0, 5], [1, -1, -1, -1, 1]], dtype=np.float32)
b4d = np.array([7, 2, -5, 6,-3], dtype=np.float32)

sol_4d, swaps_4d = gauss_jordan(A4d, b4d, dtype=np.float32)
print("Solución 7d:", sol_4d)
print("Intercambios de fila 7d:", swaps_4d)
```
